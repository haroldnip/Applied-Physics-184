```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# List of file paths for your CSV files
file_paths = ['App Physics 184 _ Rotation Sensor.csv']

# Initialize lists to store data
resistance_data = []
angle_data = []

# Read each file and append data
for file in file_paths:
    data = pd.read_csv(file)
    resistance_data.extend(data['Potentiometer Value'])  # Use extend to flatten the list
    angle_data.extend(data['Angle'])

# Convert lists to DataFrames for easier calculations
resistance_df = pd.DataFrame(resistance_data, columns=["Resistance"])
angle_df = pd.DataFrame(angle_data, columns=["Angle"])

# Reverse both resistance and angle data if needed
resistance_df = resistance_df.iloc[::-1].reset_index(drop=True)
#angle_df = angle_df.iloc[::-1].reset_index(drop=True)

# Calculate linear regression to get the R^2 value
slope, intercept, r_value, p_value, std_err = linregress(resistance_df["Resistance"], angle_df["Angle"])
r_squared = r_value**2

# Plot Position vs Time
plt.figure(figsize=(10, 5))
plt.plot(resistance_df, angle_df, label=f"RÂ² = {r_squared:.3f}")
plt.xlabel('Resistance [Î©]', size=20)
plt.ylabel('Angle [degrees]', size=20)
plt.grid(True)
plt.legend(fontsize=13)
#plt.title(f'Calibration Curve (RÂ² = {r_squared:.3f})')
plt.savefig('Reversedddd Calibration Curve')
plt.show()

print(f"RÂ² value: {r_squared:.3f}")
```
