```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the CSV file
file_path = 'd8.csv'  # Replace with your file's path
data = pd.read_csv(file_path)

# Inspect the first few rows to verify the structure
print(data.head())

# Extract Time and Voltage columns
time = data['Time']  # Replace with the actual column name for time
voltage = data['Volt']  # Replace with the actual column name for voltage

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(time, voltage, marker='o', linestyle='-', color='blue', label='Voltage')

# Customize the plot
#plt.title('Voltage vs Time')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.ylim(-0.001, 0.0015)
plt.grid(True)
plt.legend()
plt.savefig("d8")
# Show the plot
plt.show()
```
