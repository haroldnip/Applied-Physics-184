```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
signal_length = 1000       # Number of data points in the signal
threshold = 1.0            # Threshold value
noise_level = 0.5          # Noise standard deviation

# Generate a clean signal (e.g., sinusoidal)
x = np.linspace(0, 10, signal_length)
clean_signal = np.sin(2 * np.pi * x)

# Add Gaussian noise to the signal
noisy_signal = clean_signal + np.random.normal(0, noise_level, signal_length)

# Detect values above the threshold
detected_indices = np.where(noisy_signal > threshold)[0]
detected_values = noisy_signal[detected_indices]

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(x, noisy_signal, label="Noisy Signal", alpha=0.7)
plt.plot(x, clean_signal, label="Clean Signal", linestyle="--", alpha=0.7)
plt.axhline(threshold, color='red', linestyle='--', label="Threshold")
plt.scatter(x[detected_indices], detected_values, color='orange', label="Detected Points")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal Detection with Noise")
plt.legend()
plt.grid()
plt.show()

# Print detected values and their indices
print("Detected Indices:", detected_indices)
print("Detected Values:", detected_values)
```
