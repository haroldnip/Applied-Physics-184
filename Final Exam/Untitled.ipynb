```python
import numpy as np
import matplotlib.pyplot as plt
```

First Frequency

```python
f = 441 #linear frequency
w = 2. * np.pi * f #angular frequency
time_interval = 100
samples = 5000
t = np.linspace(0, time_interval, samples)
A = float(10*np.log(76)) #amplitude
y1 = A*np.sin(w * t)
plt.plot(t, y1)
plt.show()
```

```python
def exp(t, A, lbda):
    r"""y(t) = A \cdot \exp(-\lambda t)"""
    return A * np.exp(-lbda * t)
```

```python
plt.plot(np.linspace(0,10), exp(1, 1, 1))
```
