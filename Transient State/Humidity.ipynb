```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "Humidity.csv"  # Replace with your file path
data = pd.read_csv(file_path)

# Ensure the columns are named "Time" and "Data"
time = data["Time [s]"]
values = data["Humidity"]
time_numeric = [float(t) for t in time]
# Plot the data
plt.figure(figsize=(20, 12))
plt.plot(time, values, label='Data', color='blue', linewidth=2)
plt.xticks(ticks=range(0, max(time_numeric) + 1, 2), rotation=45)# Add labels and title
plt.xlabel("Time [s] ", fontsize=14)
plt.ylabel("Humidity [%]", fontsize=14)
#plt.title("Time vs Data Plot", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "hum and temp.csv"  # Replace with your file path
data = pd.read_csv(file_path)

# Ensure the columns are named "Time" and "Data"
time = data["Time [s]"]
values = data["Humidity"]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, values, label='Data', color='blue', linewidth=2)

# Add labels and title
plt.xlabel("Time [s]", fontsize=20)
plt.ylabel("Humidity [%]", fontsize=20)
#plt.title("Time vs Data Plot", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
#plt.legend()
#plt.tight_layout()

# Show the plot
plt.show()


from matplotlib.ticker import MaxNLocator

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, values, label='Data', color='blue', linewidth=2)

# Customize ticks
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True, nbins=10))  # Adjust the number of ticks

# Add labels and title
plt.xlabel("Time [s]", fontsize=20)
plt.ylabel("Humidity [%]", fontsize=20)
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.savefig("Humidity")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "hum and temp.csv"  # Replace with your file path
data = pd.read_csv(file_path)

# Ensure the columns are named "Time" and "Data"
time = data["Time [s]"]
values = data["Temperature"]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, values, label='Data', color='blue', linewidth=2)

# Add labels and title
plt.xlabel("Time [s]", fontsize=20)
plt.ylabel("Temperature", fontsize=20)
#plt.title("Time vs Data Plot", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
#plt.legend()
plt.tight_layout()

# Show the plot
plt.show()

from matplotlib.ticker import MaxNLocator

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, values, label='Data', color='blue', linewidth=2)

# Customize ticks
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True, nbins=10))  # Adjust the number of ticks

# Add labels and title
plt.xlabel("Time [s]", fontsize=20)
plt.ylabel("Temperature [$\degree$ $C$]", fontsize=20)
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.savefig("Temperature")
plt.show()
```
