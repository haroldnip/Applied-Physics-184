```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "LDR2.csv"  # Replace with your file path
data = pd.read_csv(file_path)

# Ensure the columns are named "Time" and "Data"
time = data["Time [s]"]
values = data["Reading"]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, values, label='Data', color='blue', linewidth=2)

# Add labels and title
plt.xlabel("Time [s]", fontsize=20)
plt.ylabel("Analog Reading (Resistance)", fontsize=20)
#plt.title("Time vs Data Plot", fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
#plt.legend()
plt.tight_layout()

# Show the plot
plt.savefig("Photoresistor")
plt.show()
```
